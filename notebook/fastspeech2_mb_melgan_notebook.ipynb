# TensorflowTTS(FastSpeech2 + MB-MelGAN)中文转语音

## 导入依赖

```python
from tts import fastspeech2_mb_melgan, predicted_mel_after_spectrogram

import IPython.display as ipd
```

## 输入文字

```python
input_text = "这是一个开源的端到端中文语音合成系统"
```

## 合成语音

```python
mels, audios = fastspeech2_mb_melgan.do_synthesis(input_text)
predicted_mel_after_spectrogram.do_visualize(mels[0])
ipd.Audio(audios, rate=24000)
```
