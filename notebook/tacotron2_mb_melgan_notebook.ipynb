# TensorflowTTS(Tacotron2 + MB-MelGAN)中文转语音

## 导入依赖

```python
from tts import tacotron2_mb_melgan, alignment_steps, predicted_mel_after_spectrogram

import IPython.display as ipd
```

## 输入文字

```python
input_text = "这是一个开源的端到端中文语音合成系统"
```

## 合成语音

```python
# 如果想尝试tacotron2那就打开安装窗口
tacotron2_mb_melgan.tacotron2.setup_window(win_front=5, win_back=5)

mels, alignment_history, audios = tacotron2_mb_melgan.do_synthesis(input_text)
alignment_steps.do_visualize(alignment_history[0])
predicted_mel_after_spectrogram.do_visualize(mels[0])
ipd.Audio(audios, rate=24000)
```
